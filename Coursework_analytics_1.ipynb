```python
# импортируем библиотеки

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# выгружаем данные в датафреймы

courses = pd.read_csv('Data_2/courses.csv', index_col=0)
course_contents = pd.read_csv('Data_2/course_contents.csv')
progresses = pd.read_csv('Data_2/progresses.csv')
progress_phases = pd.read_csv('Data_2/progress_phases.csv')

# объединяем датафреймы в один

main = progress_phases.merge(progresses, how='outer', left_on='progress_id', right_on='id').drop('id', axis=1)
main = main.merge(courses, how='outer', left_on='course_id', right_on='id').drop('id', axis=1)
main = main.merge(course_contents, how='outer', on=['course_id', 'module_number', 'lesson_number'])
```

```python
main[main.student_id == '768c2987a744c51ce64a5993a2a94eaf']
```

```python
main.head()
```

```python
# cчитаем общее количество курсов в датасете

total_courses = len(list(main.groupby('course_id', as_index=False)['course_id']))
print(f'Общее количество курсов: {total_courses}')
```

```python
# считаем количество модулей на каждом курсе

modules = main.groupby('title')['module_title'].nunique().reset_index().sort_values('module_title', ascending=False)
modules
```

```python
# считаем количество уроков в каждом модуле на каждом курсе

lessons = main.groupby(['title', 'module_title'])['lesson_title'].nunique().reset_index().sort_values('lesson_title')

lessons
```

```python
# считаем медианное количество уроков в модуле на каждом курсе

lessons_median = lessons.groupby('title')['lesson_title'].median().reset_index().sort_values('lesson_title')
lessons_median
```

```python
# считаем количество учеников на каждом курсе

students_count = main.groupby('title', as_index=False)['student_id'].nunique().sort_values('student_id', ascending=False)
students_count
```

```python
# вычисляем возраст студентов в датафрейме students

students = pd.read_csv('Data_2/students.csv', parse_dates=['birthday'], dayfirst=False)
```

```python
import datetime as dt
from math import trunc

students['birthday'] = pd.to_datetime(students['birthday'], errors='coerce')
students['days'] = dt.datetime.today() - students['birthday']
students['age'] = students['days'].apply(lambda x: x.days / 365.2425)
```

```python
students = students.dropna(subset = ['age'])
```

```python
students['age'] = students['age'].apply(int)
```

```python
students
```

```python
# отбираем студентов с корректным возрастом (14-80 лет)

students = students[(students['age'] > 14) & (students['age'] < 80)]
students
```

```python
# вычисляем минимальный, максимальный, средний, медианный возраст студентов

min_age = students['age'].min()
max_age = students['age'].max()
mean_age = round(students['age'].mean(), 2)
median_age = students['age'].median()

print(f'Минимальный возраст: {min_age}, максимальный: {max_age}, cредний: {mean_age}, медианный: {median_age}')
```

```python
# объединим датафреймы students и main в датафрейм main_merged

main_merged = main.merge(students, how='outer', left_on='student_id', right_on='id').drop(['id_', 'id', 'days'], axis=1)
```

```python
# минимальный, максимальный, средний, медианный возраст студентов на каждом курсе

courses_age = main_merged.groupby('title')['age'].agg(['min', 'max', 'mean', 'median']).sort_values('median')
courses_age
```

```python
# cтроим столбиковую диаграмму, отражающую количество студентов на каждом курсе

data = main.groupby('title', as_index=False)['student_id'].nunique().sort_values('student_id', ascending=False)
fig, ax = plt.subplots(figsize=(12,6))
ax.bar(data['title'], data['student_id'], color='blue')
ax.set_title('Количество студентов на курсах')
ax.set_xlabel('Курс')
ax.set_ylabel('Количество, чел.')
ax.tick_params(axis='x', rotation=90)

plt.show()
```

```python
# cтроим столбиковую диаграмму, отражающую количество студентов на каждом курсе

data = main.groupby('title', as_index=False).agg({'student_id': pd.Series.nunique, 'field': 'first'}).sort_values(['field','student_id'])
fig, ax = plt.subplots(figsize=(12,8))
sns.barplot(y=data['title'], x=data['student_id'], alpha=0.3, hue=data['field'], dodge=False)
ax.axvline(x=data['student_id'].median(), color='grey', linewidth=1, linestyle='--', label='Median')
ax.set_title('Количество студентов на курсах')
ax.set_xlabel('Количество, чел.')
ax.set_ylabel('Курс')

ax.legend()

plt.box(on=None)
plt.show()
```
